```python
#The first five rows of the data does show that several columns have missing values.
```

```python
#number of missing data points per column
missing_values = df_permit.isna().sum()
missing_values
```

```python
# how many total missing values do we have?
total_cells = np.product(df_permit.shape)
total_missing = missing_values.sum()
```

```python
print("Total cells: ",total_cells)
print("Total Missing Values: ",total_missing)
```

```python
# percent of data that is missing
missing_percent = (total_missing/total_cells)*100
print("Percentage of missing values is: ",missing_percent)
```

```python
# In Street Number Suffix column we can assume that the street number suffix doesn't exist is the reason for missing values in the particular column
# In Zipcode Column we can assume that zipcode wasn't recorded which is why there are missing values in that column
print("Street Number Suffix: ",missing_values["Street Number Suffix"])
print("Zipcode: ",missing_values["Zipcode"])
```
